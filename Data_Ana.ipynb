```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import OneHotEncoder
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
```

```python
df = pd.read_csv('Loan_default.csv')
```

```python
df.head(3)
```

```python
data = df.copy()
```

```python
data.drop(['LoanID'], axis=1, inplace=True)
```

```python
data.head(3)
```

```python
encoder = OneHotEncoder(sparse_output=False)
```

```python
encoded = encoder.fit_transform(data.select_dtypes(include=['object']).values)
```

```python
proces_data = pd.DataFrame(encoded, columns=encoder.get_feature_names_out(data.select_dtypes(include=['object']).columns))
```

```python
proces_data.head(3)
```

```python
data.head(3)
```

```python
result_data = pd.concat([data, proces_data], axis=1)
```

```python
result_data.head(3)
```

```python
result_data.columns
```

```python
remove_columns = data.select_dtypes(include=['object']).columns
result_data.drop(remove_columns, axis=1, inplace=True)
```

```python
result_data.head(3)
```

```python
pre_data = result_data['Default'].copy()
```

```python
result_data.drop(['Default'], axis=1, inplace=True)
```

```python
result_data.head(3)
```

Data was ready for ML (result_data, pre_data)

```python
X_train, X_test, y_train, y_test = train_test_split(
    result_data, pre_data, test_size=0.2, random_state=42)
```

```python
model = RandomForestClassifier()
```

```python
model.fit(X_train.values,y_train.values)
```

```python
y_pred = model.predict(X_test.values)
```

```python
accuracy = (accuracy_score(y_test, y_pred))*100
```
